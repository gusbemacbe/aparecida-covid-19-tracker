```python
# ð¬ð§ Libraries
# ð§ð· Bibliotecas
import datetime
import altair as alt
import operator
import pandas as pd

s = pd.read_csv('../../data/aparecida-small-sample.csv', parse_dates=['date'])

city = s[s['city'] == 'Aparecida']

base = alt.Chart(city).mark_line().encode(x = 'date').properties(width = 500)

confirmed = alt.value("#106466")
death = alt.value("#D8B08C")
recovered = alt.value("#87C232")

base.encode(y = "totalCases", color = confirmed).properties(title = "Total confirmed") | base.encode(y = "totalDeaths", color = death).properties(title = "Total deaths")
```

```python
city['new cases'] = city['totalCases'].diff()
city['new deaths'] = city['totalDeaths'].diff()
city['new recovered'] = city['totalRecovered'].diff()

# Initiate chart with data
base_cases = alt.Chart(city).mark_bar().encode(alt.X('date:T'), alt.Y('new cases:Q')).properties(width = 400)
base_deaths = alt.Chart(city).mark_bar().encode(alt.X('date:T'), alt.Y('new deaths:Q')).properties(width = 400)
base_recovered = alt.Chart(city).mark_bar().encode(alt.X('date:T'), alt.Y('new recovered:Q')).properties(width = 400)

# Load the chart
base_cases.encode(color = confirmed).properties(title = "Daily new cases") | base_deaths.encode(color = death).properties(title = "Daily new deaths") | base_recovered.encode(color = recovered).properties(title = "Daily new recovered")
```

```python
city['total_cases'] = city['totalCases']
city['total_deaths'] = city['totalDeaths']
city['total_recovered'] = city['totalRecovered']

tempTotalCases = city[['date','total_cases']]
tempTotalCases["title"] = "Confirmed"

tempTotalDeaths = city[['date','total_deaths']]
tempTotalDeaths["title"] = "Deaths"

tempTotalRecovered = city[['date','total_recovered']]
tempTotalRecovered["title"] = "Recovered"

temp = tempTotalCases.append(tempTotalDeaths)
temp = temp.append(tempTotalRecovered)

totalCases = alt.Chart(temp).mark_bar().encode(alt.X('date:T', title = None), alt.Y('total_cases:Q', title = None)) # color='#106466' 
totalDeaths = alt.Chart(temp).mark_bar().encode(alt.X('date:T', title = None), alt.Y('total_deaths:Q', title = None)) # color = '#DC143C'
totalRecovered = alt.Chart(temp).mark_bar().encode(alt.X('date:T', title = None), alt.Y('total_recovered:Q', title = None)) # color = '#87C232'

(totalCases + totalRecovered + totalDeaths).encode(color=alt.Color('title', scale = alt.Scale(range = ['#106466','#DC143C','#87C232']), legend = alt.Legend(title="Legend colour"))).properties(title = "Cumulative number of confirmed cases, deaths and recovered", width = 800)
```

```python
city['daily_cases'] = city['totalCases'].diff()
city['daily_deaths'] = city['totalDeaths'].diff()
city['daily_recovered'] = city['totalRecovered'].diff()

# Initiate chart with data
cases = alt.Chart(city).mark_bar(color = '#106466', opacity = 0.5).encode(alt.X('date:T'), alt.Y('daily_cases:Q', title = 'Daily new cases'))
deaths = alt.Chart(city).mark_bar(color = '#DC143C', opacity = 0.5).encode(alt.X('date:T'), alt.Y('daily_deaths:Q', title = 'Daily new deaths'))
recovered = alt.Chart(city).mark_bar(color = '#87C232', opacity = 0.5).encode(alt.X('date:T'), alt.Y('daily_recovered:Q', title = 'Daily new recovered'))

(cases + recovered + deaths).properties(title = "Daily new tolls", width = 600)
```
